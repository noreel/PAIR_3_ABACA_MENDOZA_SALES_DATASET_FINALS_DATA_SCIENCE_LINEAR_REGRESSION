#Modelling

##Linear Regression

```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

X = df['Price']  # Independent variable
y = df['Quantity Sold']  # Dependent variable

X = sm.add_constant(X)

model = sm.OLS(y, X).fit()

print(model.summary())

y_pred = model.predict(X)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='Price', y='Quantity Sold', data=df, label='Data')
plt.plot(df['Price'], y_pred, color='red', label='Regression Line')
plt.xlabel('Price')
plt.ylabel('Quantity Sold')
plt.title('Linear Regression: Price vs. Quantity Sold')
plt.legend()
plt.show()
```

##Is the Model a Good Fit?
Yes, the model appears to be a good fit. The R-squared value is close to 1, which indicates that the model explains a large portion of the variance in the dependent variable.
##What's the relationship between price and quantity?
Generally, as the price goes up, the quantity demanded goes down. This is because higher prices discourage buying, while lower prices encourage it.
